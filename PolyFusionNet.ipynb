```python
import os
import random
import csv
import cv2
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader, Dataset
from torchvision import models, transforms
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, precision_score, recall_score)
from tqdm import tqdm
from typing import Tuple, Dict, List, Optional, Union, Any
from datetime import datetime
from pathlib import Path

IMAGE_SIZE = (256, 256)
BATCH_SIZE = 16
TEST_SIZE = 0.2
VAL_SIZE = 0.1
LEARNING_RATE = 0.001
NUM_EPOCHS = 50
SEG_IMAGE_DIR = Path('data/CVC/PNG/Original')
SEG_ANNOTATION_DIR = Path('data/CVC/PNG/Ground Truth')
SSL_IMAGE_DIR = Path('data/hyper-kvasir/unlabeled-images/images')
SSL_BATCH_SIZE = 32
SSL_PRETRAIN_EPOCHS = 50
SSL_LEARNING_RATE = 0.0001
OUTPUT_DIR = Path('ablation_results10')
EARLY_STOPPING_PATIENCE = 5
HYPERKVASIR_PERCENTAGE = 0.2
RANDOM_SEED = 42
# ===== Config Flags =====
RUN_ABLATION_STUDY = True 
RUN_FULL_TRAINING = True 

random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
torch.manual_seed(RANDOM_SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(RANDOM_SEED)

OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

def generate_unique_filename(base: str, run_label: str, ext: str) -> Path:
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    return OUTPUT_DIR / f"{base}_{run_label}_{timestamp}.{ext}"

def setup_device() -> torch.device:
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    if device.type == 'cuda':
        print(f"Using GPU: {torch.cuda.get_device_name(0)}")
        torch.backends.cudnn.benchmark = True
    else:
        print("GPU not available, using CPU.")
    return device

def get_total_unlabeled_images(img_dir: Union[str, Path]) -> int:
    img_dir = Path(img_dir)
    valid_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.tif', '.tiff')
    return len([f for f in os.listdir(img_dir) if f.lower().endswith(valid_extensions)])

class ImagePreprocessor:
    @staticmethod
    def canny_edge_detection(image: np.ndarray) -> np.ndarray:
        if image.ndim == 3 and image.shape[2] == 3:
            image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        if image.dtype != np.uint8:
            if image.max() <= 1.0:
                image = (image * 255).astype(np.uint8)
            else:
                image = image.astype(np.uint8)
        return cv2.Canny(image, threshold1=100, threshold2=200)

    @staticmethod
    def sobel_edge_detection(image: np.ndarray) -> np.ndarray:
         if image.ndim == 3 and image.shape[2] == 3:
            image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
         if image.dtype == np.uint8:
             image = image.astype(np.float32) / 255.0
         grad_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
         grad_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
         magnitude = cv2.magnitude(grad_x, grad_y)
         return cv2.normalize(magnitude, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

    @staticmethod
    def morphological_operations(image: np.ndarray) -> np.ndarray:
        kernel = np.ones((5,5), np.uint8)
        if image.dtype != np.uint8:
             if image.max() <= 1.0 and image.min() >= 0:
                 image = (image * 255).astype(np.uint8)
             else:
                 image = np.clip(image, 0, 255).astype(np.uint8)
        return cv2.dilate(image, kernel, iterations=1)

    @staticmethod
    def hsv_thresholding(image: np.ndarray) -> np.ndarray:
        if image.dtype != np.uint8:
             if image.max() <= 1.0 and image.min() >= 0:
                 image = (image * 255).astype(np.uint8)
             else:
                 image = np.clip(image, 0, 255).astype(np.uint8)
        if image.ndim == 2 or image.shape[2] == 1:
            image = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
        hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
        lower_bound = np.array([30, 50, 50])
        upper_bound = np.array([90, 255, 255])
        return cv2.inRange(hsv, lower_bound, upper_bound)
    @staticmethod
    def generate_all_classical_features(image: np.ndarray) -> Dict[str, np.ndarray]:
        """Returns a dict of all classical features for ablation."""
        features = {}
        features["canny"] = ImagePreprocessor.canny_edge_detection(image)
        features["sobel"] = ImagePreprocessor.sobel_edge_detection(image)
        features["hsv"] = ImagePreprocessor.hsv_thresholding(image)
        return features
class DataLoaderHelper:
    @staticmethod
    def list_files(img_dir: Union[str, Path], annotation_dir: Optional[Union[str, Path]] = None) -> Tuple[List[Path], Optional[List[Path]]]:
        img_dir = Path(img_dir)
        valid_extensions = ('.png', '.jpg', '.jpeg', '.bmp', '.tif', '.tiff')
        image_files = sorted([p for p in img_dir.glob('*') if p.suffix.lower() in valid_extensions])
        if annotation_dir:
            annotation_dir = Path(annotation_dir)
            annotation_files = sorted([p for p in annotation_dir.glob('*') if p.suffix.lower() in valid_extensions])
            if len(image_files) != len(annotation_files):
                print(f"Warning: Mismatch in number of images ({len(image_files)}) and annotations ({len(annotation_files)}). Check file naming.")
                ann_map = {f.stem: f for f in annotation_files}
                matched_anns = [ann_map.get(img.stem) for img in image_files]
                filtered_images = [img for img, ann in zip(image_files, matched_anns) if ann is not None]
                matched_anns = [ann for ann in matched_anns if ann is not None]
                print(f"Found {len(matched_anns)} matching pairs based on filename stem.")
                if not matched_anns:
                     raise ValueError("Could not match any image/annotation pairs.")
                return filtered_images, matched_anns
            return image_files, annotation_files
        else:
            return image_files, None

    @staticmethod
    def generate_classical_features_for_list(
        image_paths: List[Path], 
        feature_type: str = "canny"  # Options: "canny", "sobel", "hsv"
    ) -> np.ndarray:
        features = []
        for img_path in tqdm(image_paths, desc=f"Generating {feature_type} features"):
            img = cv2.imread(str(img_path))
            if img is None:
                features.append(np.zeros(IMAGE_SIZE, dtype=np.float32))
                continue
            img = cv2.resize(img, IMAGE_SIZE)
            feature_dict = ImagePreprocessor.generate_all_classical_features(img)
            features.append(feature_dict[feature_type].astype(np.float32) / 255.0)
        return np.expand_dims(np.array(features), axis=1)  # Shape: (N, 1, H, W)

class SegmentationDataset(Dataset):
    def __init__(self, image_paths: List[Path], mask_paths: List[Path],
                 classical_features: Optional[np.ndarray] = None,
                 size: Tuple[int, int] = IMAGE_SIZE, transform: Optional[Any] = None):
        self.image_paths = image_paths
        self.mask_paths = mask_paths
        self.classical_features = classical_features
        self.size = size
        self.transform = transform
        if classical_features is not None and len(image_paths) != len(classical_features):
            raise ValueError("Mismatch between number of images and classical features provided.")
        if len(image_paths) != len(mask_paths):
             raise ValueError("Mismatch between number of images and masks provided.")

    def __len__(self) -> int:
        return len(self.image_paths)

    def __getitem__(self, idx: int) -> Union[Tuple[torch.Tensor, torch.Tensor], Tuple[torch.Tensor, torch.Tensor, torch.Tensor]]:
        img_path = self.image_paths[idx]
        mask_path = self.mask_paths[idx]
        image = cv2.imread(str(img_path))
        if image is None:
            raise IOError(f"Could not read image: {img_path}")
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image = cv2.resize(image, self.size, interpolation=cv2.INTER_LINEAR)
        mask = cv2.imread(str(mask_path), cv2.IMREAD_GRAYSCALE)
        if mask is None:
            raise IOError(f"Could not read mask: {mask_path}")
        mask = cv2.resize(mask, self.size, interpolation=cv2.INTER_NEAREST)
        mask = (mask > 0).astype(np.float32)
        if self.transform:
             if random.random() > 0.5:
                 image = np.fliplr(image).copy()
                 mask = np.fliplr(mask).copy()
        image = np.transpose(image, (2, 0, 1)).astype(np.float32) / 255.0
        mask = np.expand_dims(mask, axis=0)
        image_tensor = torch.from_numpy(image)
        mask_tensor = torch.from_numpy(mask)
        if self.classical_features is not None:
            classical_feat = self.classical_features[idx]
            if self.transform and 'fliplr' in str(self.transform).lower():
                 if random.random() > 0.5:
                      classical_feat = np.fliplr(classical_feat).copy()
            classical_tensor = torch.from_numpy(classical_feat)
            return image_tensor, classical_tensor, mask_tensor
        else:
            return image_tensor, mask_tensor

class SelfSupervisedDataset(Dataset):
    def __init__(self, img_dir: Union[str, Path], transform: Optional[transforms.Compose] = None, max_percentage: Optional[float] = None):
        self.img_dir = Path(img_dir)
        self.transform = transform or self._get_default_transform()
        valid_extensions = ('.jpg', '.jpeg', '.png')
        self.image_files = [p for p in self.img_dir.glob('*') if p.suffix.lower() in valid_extensions]
        if max_percentage is not None and 0 < max_percentage <= 1:
            num_to_load = int(len(self.image_files) * max_percentage)
            random.shuffle(self.image_files)
            self.image_files = self.image_files[:num_to_load]
        print(f"Using {len(self.image_files)} unlabeled images for pretraining.")

    def _get_default_transform(self) -> transforms.Compose:
        return transforms.Compose([
            transforms.RandomResizedCrop(size=IMAGE_SIZE[0], scale=(0.2, 1.0)),
            transforms.RandomHorizontalFlip(),
            transforms.RandomApply([
                transforms.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4, hue=0.1)
            ], p=0.8),
            transforms.RandomGrayscale(p=0.2),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])

    def __len__(self) -> int:
        return len(self.image_files)

    def __getitem__(self, idx: int) -> Tuple[torch.Tensor, torch.Tensor]:
        img_path = self.image_files[idx]
        try:
            image_cv = cv2.imread(str(img_path))
            if image_cv is None: raise IOError(f"Cannot load {img_path}")
            image_pil = Image.fromarray(cv2.cvtColor(image_cv, cv2.COLOR_BGR2RGB))
        except Exception as e:
            print(f"Error loading image {img_path}, returning dummy data: {e}")
            dummy_tensor = torch.zeros((3, IMAGE_SIZE[0], IMAGE_SIZE[1]))
            return dummy_tensor, dummy_tensor
        aug1 = self.transform(image_pil)
        aug2 = self.transform(image_pil)
        return aug1, aug2

class ContrastiveModel(nn.Module):
    def __init__(self, base_encoder: nn.Module, projection_dim: int = 128):
        super().__init__()
        self.base_model = base_encoder
        if isinstance(base_encoder, models.ResNet):
            num_ftrs = base_encoder.fc.in_features
            base_encoder.fc = nn.Identity()
        elif isinstance(base_encoder, models.VisionTransformer):
             num_ftrs = base_encoder.heads.head.in_features
             base_encoder.heads.head = nn.Identity()
        else:
            try:
                 if hasattr(base_encoder, 'classifier') and isinstance(base_encoder.classifier, nn.Linear):
                     num_ftrs = base_encoder.classifier.in_features
                     base_encoder.classifier = nn.Identity()
                 elif hasattr(base_encoder, 'fc') and isinstance(base_encoder.fc, nn.Linear):
                     num_ftrs = base_encoder.fc.in_features
                     base_encoder.fc = nn.Identity()
                 else:
                     with torch.no_grad():
                         dummy_input = torch.randn(1, 3, *IMAGE_SIZE)
                         output = base_encoder(dummy_input)
                         num_ftrs = output.shape[-1]
                     print("Warning: Automatically detecting feature dimension. Ensure the base_encoder outputs features.")
            except Exception as e:
                 raise ValueError(f"Unsupported base encoder type or could not determine feature dimension: {e}")
        self.projection_head = nn.Sequential(
            nn.Linear(num_ftrs, projection_dim),
            nn.ReLU(),
            nn.Linear(projection_dim, projection_dim)
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        features = self.base_model(x)
        return self.projection_head(features)

class AttentionFusion(nn.Module):
    def __init__(self, in_channels: int):
        super().__init__()
        self.query_conv = nn.Conv2d(in_channels, in_channels // 8, kernel_size=1)
        self.key_conv = nn.Conv2d(in_channels, in_channels // 8, kernel_size=1)
        self.value_conv = nn.Conv2d(in_channels, in_channels, kernel_size=1)
        self.gamma = nn.Parameter(torch.zeros(1))

    def forward(self, x1: torch.Tensor, x2: Optional[torch.Tensor] = None) -> torch.Tensor:
        if x2 is None:
            combined = x1
        else:
            if x1.shape[2:] != x2.shape[2:]:
                 x2 = F.interpolate(x2, size=x1.shape[2:], mode='bilinear', align_corners=False)
            if x1.shape[1] != x2.shape[1]:
                if x1.shape[1] > x2.shape[1]:
                    print(f"Warning: Channel mismatch in AttentionFusion ({x1.shape[1]} vs {x2.shape[1]}). Using only x1.")
                    combined = x1
                else:
                     print(f"Warning: Channel mismatch in AttentionFusion ({x1.shape[1]} vs {x2.shape[1]}). Using only x1.")
                     combined = x1
            else:
                 combined = x1 + x2
        m_batchsize, C, height, width = combined.size()
        proj_query = self.query_conv(combined).view(m_batchsize, -1, width * height).permute(0, 2, 1)
        proj_key = self.key_conv(combined).view(m_batchsize, -1, width * height)
        energy = torch.bmm(proj_query, proj_key)
        attention = F.softmax(energy, dim=-1)
        proj_value = self.value_conv(combined).view(m_batchsize, -1, width * height)
        out = torch.bmm(proj_value, attention.permute(0, 2, 1))
        out = out.view(m_batchsize, C, height, width)
        return self.gamma * out + combined

class HybridPolypSegmentationModel(nn.Module):
    def __init__(self, encoder_backbone: nn.Module, use_classical: bool = False, num_classes: int = 1):
        super().__init__()
        self.use_classical = use_classical
        self.encoder = encoder_backbone
        if hasattr(self.encoder, 'fc'):
            encoder_output_channels = self.encoder.fc.in_features if isinstance(self.encoder.fc, nn.Linear) else 512
        elif hasattr(self.encoder, 'layer4') and hasattr(self.encoder.layer4[-1], 'conv3'):
             encoder_output_channels = self.encoder.layer4[-1].conv3.out_channels
        elif hasattr(self.encoder, 'layer4') and hasattr(self.encoder.layer4[-1], 'conv2'):
             encoder_output_channels = self.encoder.layer4[-1].conv2.out_channels
        else:
            print("Warning: Could not automatically determine ResNet encoder output channels. Assuming 512.")
            encoder_output_channels = 512
        self.encoder.fc = nn.Identity()
        if hasattr(self.encoder, 'avgpool'):
             self.encoder.avgpool = nn.Identity()
        classical_channels = 1 if use_classical else 0
        self.attention_fusion = AttentionFusion(encoder_output_channels)
        decoder_input_channels = encoder_output_channels + classical_channels
        self.decoder_block1 = self._make_decoder_block(decoder_input_channels, 256)
        self.decoder_block2 = self._make_decoder_block(256, 128)
        self.decoder_block3 = self._make_decoder_block(128, 64)
        self.decoder_block4 = self._make_decoder_block(64, 32)
        self.decoder_block5 = self._make_decoder_block(32, 16)
        self.final_conv = nn.Conv2d(16, num_classes, kernel_size=1)

    def _make_decoder_block(self, in_channels: int, out_channels: int) -> nn.Sequential:
        return nn.Sequential(
            nn.ConvTranspose2d(in_channels, out_channels, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
            nn.Dropout2d(0.2)
        )
    def forward(self, x: torch.Tensor, classical_feat: Optional[torch.Tensor] = None) -> torch.Tensor:
        if x.shape[1] == 1:
             x = x.repeat(1, 3, 1, 1)
        e1 = self.encoder.conv1(x)
        e1 = self.encoder.bn1(e1)
        e1 = self.encoder.relu(e1)
        e1 = self.encoder.maxpool(e1)
        e2 = self.encoder.layer1(e1)
        e3 = self.encoder.layer2(e2)
        e4 = self.encoder.layer3(e3)
        encoder_features = self.encoder.layer4(e4)
        attended_features = self.attention_fusion(encoder_features)
        decoder_input = attended_features
        if self.use_classical and classical_feat is not None:
            target_h, target_w = attended_features.shape[2], attended_features.shape[3]
            classical_feat_resized = F.interpolate(classical_feat, size=(target_h, target_w),
                                                   mode='bilinear', align_corners=False)
            decoder_input = torch.cat([attended_features, classical_feat_resized], dim=1)
        elif self.use_classical and classical_feat is None:
             print("Warning: Model configured to use classical features, but none provided in forward pass.")
        d1 = self.decoder_block1(decoder_input)
        d2 = self.decoder_block2(d1)
        d3 = self.decoder_block3(d2)
        d4 = self.decoder_block4(d3)
        d5 = self.decoder_block5(d4)
        output = self.final_conv(d5)
        if output.shape[2:] != IMAGE_SIZE:
            output = F.interpolate(output, size=IMAGE_SIZE, mode='bilinear', align_corners=False)
        return output

def dice_coefficient(preds: torch.Tensor, targets: torch.Tensor, smooth: float = 1e-6) -> torch.Tensor:
    preds = torch.sigmoid(preds)
    preds = (preds > 0.5).float()
    if preds.shape != targets.shape:
        raise ValueError(f"Shape mismatch: preds {preds.shape}, targets {targets.shape}")
    preds_flat = preds.contiguous().view(preds.shape[0], -1)
    targets_flat = targets.contiguous().view(targets.shape[0], -1)
    intersection = (preds_flat * targets_flat).sum(dim=1)
    pred_sum = preds_flat.sum(dim=1)
    target_sum = targets_flat.sum(dim=1)
    dice = (2. * intersection + smooth) / (pred_sum + target_sum + smooth)
    return dice.mean()

def iou_score(preds: torch.Tensor, targets: torch.Tensor, smooth: float = 1e-6) -> torch.Tensor:
    preds = torch.sigmoid(preds)
    preds = (preds > 0.5).float()
    if preds.shape != targets.shape:
        raise ValueError(f"Shape mismatch: preds {preds.shape}, targets {targets.shape}")
    preds_flat = preds.contiguous().view(preds.shape[0], -1)
    targets_flat = targets.contiguous().view(targets.shape[0], -1)
    intersection = (preds_flat * targets_flat).sum(dim=1)
    union = preds_flat.sum(dim=1) + targets_flat.sum(dim=1) - intersection
    iou = (intersection + smooth) / (union + smooth)
    return iou.mean()

class Trainer:
    def __init__(self, model: nn.Module, device: torch.device, model_save_path: Path, use_classical: bool):
        self.model = model.to(device)
        self.device = device
        self.best_val_metric = -float('inf')
        self.early_stopping_counter = 0
        self.model_save_path = model_save_path
        self.use_classical = use_classical

    def train_epoch(self, train_loader: DataLoader, optimizer: optim.Optimizer,
                      criterion: nn.Module) -> Dict[str, float]:
        self.model.train()
        total_loss = 0.0
        total_dice = 0.0
        all_preds_flat = []
        all_targets_flat = []
        pbar = tqdm(train_loader, desc="Training")
        for batch in pbar:
            if self.use_classical:
                images, classical_feats, targets = batch
                images, classical_feats, targets = images.to(self.device), classical_feats.to(self.device), targets.to(self.device)
                outputs = self.model(images, classical_feats)
            else:
                images, targets = batch
                images, targets = images.to(self.device), targets.to(self.device)
                outputs = self.model(images)
            loss = criterion(outputs, targets)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            total_loss += loss.item() * images.size(0)
            with torch.no_grad():
                preds_binary = (torch.sigmoid(outputs) > 0.5).int()
                dice = dice_coefficient(outputs, targets)
                total_dice += dice.item() * images.size(0)
                all_preds_flat.extend(preds_binary.cpu().numpy().flatten())
                all_targets_flat.extend(targets.int().cpu().numpy().flatten())
            pbar.set_postfix(loss=loss.item(), dice=dice.item())
        avg_loss = total_loss / len(train_loader.dataset)
        avg_dice = total_dice / len(train_loader.dataset)
        avg_iou = iou_score(outputs, targets).item()
        accuracy = accuracy_score(all_targets_flat, all_preds_flat)
        precision = precision_score(all_targets_flat, all_preds_flat, zero_division=0)
        recall = recall_score(all_targets_flat, all_preds_flat, zero_division=0)
        return {
            'loss': avg_loss,
            'dice': avg_dice,
            'iou': avg_iou,
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall
        }

    def validate_epoch(self, val_loader: DataLoader, criterion: nn.Module) -> Dict[str, float]:
        self.model.eval()
        total_loss = 0.0
        total_dice = 0.0
        total_iou = 0.0
        all_preds_flat = []
        all_targets_flat = []
        pbar = tqdm(val_loader, desc="Validating")
        with torch.no_grad():
            for batch in pbar:
                if self.use_classical:
                    images, classical_feats, targets = batch
                    images, classical_feats, targets = images.to(self.device), classical_feats.to(self.device), targets.to(self.device)
                    outputs = self.model(images, classical_feats)
                else:
                    images, targets = batch
                    images, targets = images.to(self.device), targets.to(self.device)
                    outputs = self.model(images)
                loss = criterion(outputs, targets)
                total_loss += loss.item() * images.size(0)
                preds_binary = (torch.sigmoid(outputs) > 0.5).int()
                dice = dice_coefficient(outputs, targets)
                total_dice += dice.item() * images.size(0)
                total_iou = iou_score(outputs, targets).item()
                all_preds_flat.extend(preds_binary.cpu().numpy().flatten())
                all_targets_flat.extend(targets.int().cpu().numpy().flatten())
                pbar.set_postfix(loss=loss.item(), dice=dice.item())
        avg_loss = total_loss / len(val_loader.dataset)
        avg_dice = total_dice / len(val_loader.dataset)
        avg_iou = total_iou / len(val_loader.dataset)
        accuracy = accuracy_score(all_targets_flat, all_preds_flat)
        precision = precision_score(all_targets_flat, all_preds_flat, zero_division=0)
        recall = recall_score(all_targets_flat, all_preds_flat, zero_division=0)
        current_val_metric = avg_dice
        if current_val_metric > self.best_val_metric:
            print(f"Validation Dice improved ({self.best_val_metric:.4f} --> {current_val_metric:.4f}). Saving model...")
            self.best_val_metric = current_val_metric
            self.early_stopping_counter = 0
            torch.save(self.model.state_dict(), self.model_save_path)
            print(f"Model saved to {self.model_save_path}")
        else:
            self.early_stopping_counter += 1
            print(f"Validation Dice did not improve ({current_val_metric:.4f}). Early stopping counter: {self.early_stopping_counter}/{EARLY_STOPPING_PATIENCE}")
        return {
            'loss': avg_loss,
            'dice': avg_dice,
            'iou': avg_iou,
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall
        }

    def check_early_stopping(self) -> bool:
        return self.early_stopping_counter >= EARLY_STOPPING_PATIENCE

class Visualizer:
    @staticmethod
    def plot_metric(train_metric: List[float], val_metric: List[float], metric_name: str, epochs: int, filename: Path):
        plt.figure(figsize=(10, 6))
        plt.plot(range(1, epochs + 1), train_metric, label=f"Training {metric_name}", marker='o')
        plt.plot(range(1, epochs + 1), val_metric, label=f"Validation {metric_name}", marker='o')
        plt.xlabel('Epochs')
        plt.ylabel(metric_name)
        plt.title(f'Training and Validation {metric_name}')
        plt.legend()
        plt.grid(True)
        plt.savefig(filename)
        plt.close()
        print(f"Saved {metric_name} plot to {filename}")

    @staticmethod
    def plot_sample_predictions(input_image: np.ndarray, ground_truth: np.ndarray,
                                predicted_mask: np.ndarray, filename: Path):
        if input_image.ndim == 3 and input_image.shape[0] in [1, 3]:
             input_image = input_image.transpose(1, 2, 0)
        if input_image.max() <= 1.0 and input_image.dtype == np.float32:
             input_image = (input_image * 255).astype(np.uint8)
        else:
             input_image = input_image.astype(np.uint8)
        if input_image.ndim == 2:
             input_image = cv2.cvtColor(input_image, cv2.COLOR_GRAY2RGB)
        elif input_image.shape[2] == 1:
             input_image = cv2.cvtColor(input_image, cv2.COLOR_GRAY2RGB)
        if ground_truth.ndim == 3: ground_truth = ground_truth.squeeze()
        if predicted_mask.ndim == 3: predicted_mask = predicted_mask.squeeze()
        plt.figure(figsize=(15, 5))
        plt.subplot(1, 3, 1)
        plt.imshow(input_image)
        plt.title('Input Image')
        plt.axis('off')
        plt.subplot(1, 3, 2)
        plt.imshow(ground_truth, cmap='gray')
        plt.title('Ground Truth Mask')
        plt.axis('off')
        plt.subplot(1, 3, 3)
        plt.imshow(predicted_mask, cmap='gray')
        plt.title('Predicted Mask (Thresholded)')
        plt.axis('off')
        plt.tight_layout()
        plt.savefig(filename)
        plt.close()
        print(f"Saved sample prediction to {filename}")

    @staticmethod
    def save_metrics_to_csv(metrics: List[Dict[str, float]], filename: Path):
        if not metrics: return
        fieldnames = ["epoch"] + list(metrics[0].keys())
        with open(filename, 'w', newline='') as csvfile:
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writeheader()
            for epoch, metric_dict in enumerate(metrics, 1):
                row = {"epoch": epoch}
                row.update(metric_dict)
                writer.writerow(row)
        print(f"Training metrics saved to: {filename}")

def pretrain_contrastive(device: torch.device, run_label: str, max_percentage: float = 1.0) -> Optional[Path]:
    print(f"\n--- Starting Self-Supervised Pretraining ({run_label}) ---")
    print(f"Using {max_percentage*100:.1f}% of unlabeled data from {SSL_IMAGE_DIR}")
    base_encoder = models.resnet18(weights=None)
    contrastive_model = ContrastiveModel(base_encoder).to(device)
    ssl_dataset = SelfSupervisedDataset(SSL_IMAGE_DIR, max_percentage=max_percentage)
    if len(ssl_dataset) == 0:
        print("No unlabeled images found or loaded. Skipping pretraining.")
        return None
    ssl_dataloader = DataLoader(ssl_dataset, batch_size=SSL_BATCH_SIZE, shuffle=True,
                              pin_memory=True, num_workers=0)
    optimizer = optim.Adam(contrastive_model.parameters(), lr=SSL_LEARNING_RATE)
    criterion = nn.CrossEntropyLoss()
    temperature = 0.1
    contrastive_model.train()
    for epoch in range(SSL_PRETRAIN_EPOCHS):
        total_loss = 0.0
        pbar = tqdm(ssl_dataloader, desc=f"SSL Epoch {epoch+1}/{SSL_PRETRAIN_EPOCHS}")
        for aug1, aug2 in pbar:
            aug1, aug2 = aug1.to(device), aug2.to(device)
            proj1 = contrastive_model(aug1)
            proj2 = contrastive_model(aug2)
            proj1 = F.normalize(proj1, dim=1)
            proj2 = F.normalize(proj2, dim=1)
            similarity_matrix = torch.matmul(proj1, proj2.T) / temperature
            labels = torch.arange(proj1.shape[0], device=device)
            loss = criterion(similarity_matrix, labels)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            total_loss += loss.item() * aug1.size(0)
            pbar.set_postfix(loss=loss.item())
        avg_loss = total_loss / len(ssl_dataset)
        print(f"SSL Pretrain Epoch {epoch+1}, Avg Loss: {avg_loss:.4f}")
    if isinstance(contrastive_model.base_model, models.ResNet):
        pretrained_encoder_path = generate_unique_filename(f"pretrained_encoder_{base_encoder.__class__.__name__}", run_label, "pth")
        torch.save(contrastive_model.base_model.state_dict(), pretrained_encoder_path)
        print(f"Pretrained encoder saved to: {pretrained_encoder_path}")
        print("--- Self-Supervised Pretraining Finished ---")
        return pretrained_encoder_path
    else:
        print("Warning: Could not save pretrained encoder weights (base model type unknown).")
        print("--- Self-Supervised Pretraining Finished ---")
        return None

def run_experiment(
    config: Dict[str, Any],
    device: torch.device,
    all_image_paths: List[Path],
    all_mask_paths: List[Path],
    feature_type: str = "canny"
) -> Dict[str, float]:
    if config["use_classical"]:
        print(f"\n--- Using {feature_type.upper()} classical features ---")
        all_classical_features = DataLoaderHelper.generate_classical_features_for_list(all_image_paths, feature_type)
    run_label = config['run_label']
    use_ssl = config['use_ssl']
    use_classical = config['use_classical']
    pretrained_encoder_path_to_use = config.get('pretrained_encoder_path_to_use', None)
    all_classical_features = config.get('precomputed_classical_features', None)
    all_image_paths = config.get('all_image_paths', None)
    all_mask_paths = config.get('all_mask_paths', None)
    print(f"Running Experiment: {run_label}")
    print(f"Config: SSL Pretraining={use_ssl}, Classical Features={use_classical}")
    print(f"Encoder Path Used: {pretrained_encoder_path_to_use}")
    model_save_path = generate_unique_filename("best_model", run_label, "pth")
    output_csv_file = generate_unique_filename("training_metrics", run_label, "csv")
    loss_plot_file = generate_unique_filename("plot_loss", run_label, "png")
    dice_plot_file = generate_unique_filename("plot_dice", run_label, "png")
    iou_plot_file = generate_unique_filename("plot_iou", run_label, "png")
    acc_plot_file = generate_unique_filename("plot_accuracy", run_label, "png")
    prec_plot_file = generate_unique_filename("plot_precision", run_label, "png")
    rec_plot_file = generate_unique_filename("plot_recall", run_label, "png")
    print("\n--- Preparing Segmentation Data Splits ---")
    if all_image_paths is None or all_mask_paths is None:
        all_image_paths, all_mask_paths = DataLoaderHelper.list_files(SEG_IMAGE_DIR, SEG_ANNOTATION_DIR)
        if not all_image_paths or not all_mask_paths:
            raise FileNotFoundError("Could not find image or mask files. Check directories.")
    if use_classical and all_classical_features is None:
         print("Warning: Classical features requested but not precomputed. Attempting generation now...")
         all_classical_features = DataLoaderHelper.generate_classical_features_for_list(all_image_paths)
         if all_classical_features is None:
              raise ValueError("Failed to generate classical features required for this run.")
    indices = list(range(len(all_image_paths)))
    train_val_indices, test_indices = train_test_split(indices, test_size=TEST_SIZE, random_state=RANDOM_SEED)
    relative_val_size = VAL_SIZE / (1.0 - TEST_SIZE) if (1.0 - TEST_SIZE) > 0 else VAL_SIZE
    train_indices, val_indices = train_test_split(train_val_indices, test_size=relative_val_size, random_state=RANDOM_SEED)
    X_train_paths = [all_image_paths[i] for i in train_indices]
    y_train_paths = [all_mask_paths[i] for i in train_indices]
    X_val_paths = [all_image_paths[i] for i in val_indices]
    y_val_paths = [all_mask_paths[i] for i in val_indices]
    train_classical = all_classical_features[train_indices] if all_classical_features is not None and use_classical else None
    val_classical = all_classical_features[val_indices] if all_classical_features is not None and use_classical else None
    print(f"Train set size: {len(X_train_paths)}")
    print(f"Validation set size: {len(X_val_paths)}")
    print(f"Using classical features for train: {train_classical is not None}")
    print(f"Using classical features for val: {val_classical is not None}")
    train_dataset = SegmentationDataset(X_train_paths, y_train_paths, classical_features=train_classical)
    val_dataset = SegmentationDataset(X_val_paths, y_val_paths, classical_features=val_classical)
    num_loader_workers = 0
    print(f"Using {num_loader_workers} workers for DataLoaders.")
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True,
                              pin_memory=True, num_workers=num_loader_workers)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False,
                            pin_memory=True, num_workers=num_loader_workers)
    print("\n--- Initializing Model ---")
    use_imagenet_weights = not use_ssl
    encoder = models.resnet18(weights=models.ResNet18_Weights.DEFAULT if use_imagenet_weights else None)
    if use_ssl and pretrained_encoder_path_to_use and pretrained_encoder_path_to_use.exists():
        print(f"Attempting to load SSL weights from: {pretrained_encoder_path_to_use}")
        try:
            pretrained_dict = torch.load(pretrained_encoder_path_to_use, map_location=device)
            model_dict = encoder.state_dict()
            pretrained_dict = {k.replace('module.', ''): v for k, v in pretrained_dict.items() if k.replace('module.', '') in model_dict and model_dict[k.replace('module.', '')].shape == v.shape}
            if not pretrained_dict:
                 print("Warning: SSL Pretrained dictionary is empty after filtering. Check keys. Using scratch/ImageNet weights.")
            else:
                 print(f"Loading {len(pretrained_dict)} matching keys from SSL weights.")
                 model_dict.update(pretrained_dict)
                 encoder.load_state_dict(model_dict)
                 print("Successfully loaded SSL weights.")
        except Exception as e:
            print(f"Error loading provided SSL weights: {e}. Using scratch/ImageNet weights.")
            encoder = models.resnet18(weights=models.ResNet18_Weights.DEFAULT if use_imagenet_weights else None)
    elif use_ssl and not (pretrained_encoder_path_to_use and pretrained_encoder_path_to_use.exists()):
         print("Warning: SSL use requested, but valid weights path not provided or file missing. Using scratch/ImageNet weights.")
         encoder = models.resnet18(weights=models.ResNet18_Weights.DEFAULT if use_imagenet_weights else None)
    elif not use_ssl:
         print("Using ImageNet pretrained weights (or scratch if weights=None).")
    segmentation_model = HybridPolypSegmentationModel(encoder_backbone=encoder, use_classical=use_classical)
    segmentation_model.to(device)
    criterion = nn.BCEWithLogitsLoss()
    optimizer = optim.Adam(segmentation_model.parameters(), lr=LEARNING_RATE)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.1, patience=3, verbose=True)
    trainer = Trainer(segmentation_model, device, model_save_path, use_classical=use_classical)
    print("\n--- Starting Segmentation Training ---")
    all_metrics = []
    best_epoch_metrics = {}
    for epoch in range(1, NUM_EPOCHS + 1):
        print(f"\nEpoch {epoch}/{NUM_EPOCHS}")
        train_metrics = trainer.train_epoch(train_loader, optimizer, criterion)
        val_metrics = trainer.validate_epoch(val_loader, criterion)
        epoch_metrics = {f"train_{k}": v for k, v in train_metrics.items()}
        epoch_metrics.update({f"val_{k}": v for k, v in val_metrics.items()})
        all_metrics.append(epoch_metrics)
        if 'dice' in val_metrics and val_metrics['dice'] > trainer.best_val_metric:
             print(f"Validation Dice improved ({trainer.best_val_metric:.4f} --> {val_metrics['dice']:.4f}).")
             trainer.best_val_metric = val_metrics['dice']
             best_epoch_metrics = epoch_metrics
             best_epoch_metrics['best_epoch'] = epoch
        elif 'dice' not in val_metrics:
             print("Warning: 'dice' not found in validation metrics.")
        print(f"Train Loss: {train_metrics.get('loss', -1):.4f} | Val Loss: {val_metrics.get('loss', -1):.4f}")
        print(f"Train Dice: {train_metrics.get('dice', -1):.4f} | Val Dice: {val_metrics.get('dice', -1):.4f}")
        print(f"Train IoU:  {train_metrics.get('iou', -1):.4f} | Val IoU:  {val_metrics.get('iou', -1):.4f}")
        print(f"Train Acc:  {train_metrics.get('accuracy', -1):.4f} | Val Acc:  {val_metrics.get('accuracy', -1):.4f}")
        print(f"Train Prec: {train_metrics.get('precision', -1):.4f} | Val Prec: {val_metrics.get('precision', -1):.4f}")
        print(f"Train Rec:  {train_metrics.get('recall', -1):.4f} | Val Rec:  {val_metrics.get('recall', -1):.4f}")
        if 'dice' in val_metrics:
             scheduler.step(val_metrics['dice'])
        if trainer.check_early_stopping():
            print(f"Early stopping triggered at epoch {epoch}.")
            break
    print("\n--- Training Finished ---")
    if not all_metrics:
        print("Warning: No metrics recorded during training.")
        return {"status": "No metrics"}
    Visualizer.save_metrics_to_csv(all_metrics, output_csv_file)
    epochs_run = len(all_metrics)
    Visualizer.plot_metric([m['train_loss'] for m in all_metrics], [m['val_loss'] for m in all_metrics], "Loss", epochs_run, loss_plot_file)
    Visualizer.plot_metric([m['train_dice'] for m in all_metrics], [m['val_dice'] for m in all_metrics], "Dice Coefficient", epochs_run, dice_plot_file)
    Visualizer.plot_metric([m['train_iou'] for m in all_metrics], [m['val_iou'] for m in all_metrics], "IoU Score", epochs_run, iou_plot_file)
    Visualizer.plot_metric([m['train_accuracy'] for m in all_metrics], [m['val_accuracy'] for m in all_metrics], "Accuracy", epochs_run, acc_plot_file)
    Visualizer.plot_metric([m['train_precision'] for m in all_metrics], [m['val_precision'] for m in all_metrics], "Precision", epochs_run, prec_plot_file)
    Visualizer.plot_metric([m['train_recall'] for m in all_metrics], [m['val_recall'] for m in all_metrics], "Recall", epochs_run, rec_plot_file)
    print("\n--- Visualizing Sample Predictions from Best Model ---")
    if model_save_path.exists() and best_epoch_metrics:
         print(f"Loading best model state from epoch {best_epoch_metrics.get('best_epoch','N/A')} saved at {model_save_path}")
         best_model_state = torch.load(model_save_path, map_location=device)
         final_use_imagenet = not use_ssl
         final_encoder = models.resnet18(weights=models.ResNet18_Weights.DEFAULT if final_use_imagenet else None)
         final_model = HybridPolypSegmentationModel(encoder_backbone=final_encoder, use_classical=use_classical)
         final_model.load_state_dict(best_model_state)
         final_model.to(device)
         final_model.eval()
         num_samples_to_plot = 4
         sample_indices = random.sample(range(len(val_dataset)), min(num_samples_to_plot, len(val_dataset)))
         for i, idx in enumerate(sample_indices):
             if use_classical:
                 img_tensor, classical_tensor, mask_tensor = val_dataset[idx]
                 img_tensor = img_tensor.unsqueeze(0).to(device)
                 classical_tensor = classical_tensor.unsqueeze(0).to(device)
                 with torch.no_grad():
                     pred_logits = final_model(img_tensor, classical_tensor)
             else:
                 img_tensor, mask_tensor = val_dataset[idx]
                 img_tensor = img_tensor.unsqueeze(0).to(device)
                 with torch.no_grad():
                     pred_logits = final_model(img_tensor)
             pred_prob = torch.sigmoid(pred_logits)
             pred_mask_binary = (pred_prob.squeeze(0) > 0.5).cpu().numpy()[0]
             input_img_np = img_tensor.squeeze(0).cpu().numpy()
             gt_mask_np = mask_tensor.cpu().numpy()
             sample_pred_file = generate_unique_filename(f"sample_pred_{i}", run_label, "png")
             Visualizer.plot_sample_predictions(
                 input_img_np,
                 gt_mask_np,
                 pred_mask_binary,
                 sample_pred_file
             )
    else:
         print(f"Best model file not found ({model_save_path}) or no best metrics recorded. Skipping sample prediction visualization.")
    print(f"===== Experiment {run_label} Finished =====")
    if not best_epoch_metrics:
         return all_metrics[-1] if all_metrics else {"status": "No metrics recorded"}
    return best_epoch_metrics

def main():
    device = setup_device()
    # Load dataset paths once (used by all experiments)
    all_image_paths, all_mask_paths = DataLoaderHelper.list_files(SEG_IMAGE_DIR, SEG_ANNOTATION_DIR)
    if not all_image_paths or not all_mask_paths:
        raise FileNotFoundError("Could not find image or mask files. Check directories.")
    # ===== VISUALIZE CANNY EDGES (ADD THIS) =====
    sample_idx = 0
    image_path = all_image_paths[sample_idx]
    image = cv2.imread(str(image_path))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for display

    # Generate Canny edges
    edges = ImagePreprocessor.canny_edge_detection(image)

    # Display
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    plt.imshow(image)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1,2,2)
    plt.imshow(edges, cmap='gray')
    plt.title('Canny Edges')
    plt.axis('off')
    plt.show()
    # ===== END VISUALIZATION =====
    if RUN_FULL_TRAINING:
        print("\n" + "="*40)
        print("RUNNING FULL MODEL ABLATION STUDY")
        print("="*40)
       
        ablation_configs = [
            {"run_label": "Baseline", "use_ssl": False, "use_classical": False},
            {"run_label": "Classical", "use_ssl": False, "use_classical": True},
            {"run_label": "SSL", "use_ssl": True, "use_classical": False},
            {"run_label": "SSL_and_Classical", "use_ssl": True, "use_classical": True},
        ]
        ssl_needed = any(cfg["use_ssl"] for cfg in ablation_configs)
        pretrained_encoder_path = None
        if ssl_needed:
            print("\n--- Running SSL Pretraining ---")
            pretrained_encoder_path = pretrain_contrastive(device, "ResNet18_SSL", 
                                                         max_percentage=HYPERKVASIR_PERCENTAGE)
            if pretrained_encoder_path is None:
                print("Warning: SSL pretraining failed - SSL models will train from scratch")
        classical_needed = any(cfg["use_classical"] for cfg in ablation_configs)
        all_classical_features = None
        if classical_needed:
            print("\n--- Generating Classical Features ---")
            all_classical_features = DataLoaderHelper.generate_classical_features_for_list(
                all_image_paths, "canny"
            )
        model_ablation_results = {}
        for config in ablation_configs:
            print(f"\n{'='*20} Running: {config['run_label']} {'='*20}")
            
            full_config = {
                **config,
                "pretrained_encoder_path_to_use": pretrained_encoder_path if config["use_ssl"] else None,
                "precomputed_classical_features": all_classical_features if config["use_classical"] else None,
            }

            try:
                metrics = run_experiment(
                    config=full_config,
                    device=device,
                    all_image_paths=all_image_paths,
                    all_mask_paths=all_mask_paths,
                    feature_type="canny"
                )
                model_ablation_results[config["run_label"]] = metrics
            except Exception as e:
                print(f"Error in {config['run_label']}: {str(e)}")
                model_ablation_results[config["run_label"]] = {"error": str(e)}

        # Print final results
        print("\n\n=== Final Ablation Results ===")
        print(f"{'Model':<25} | {'Val Dice':<10} | {'Val_IoU':<10} | {'Val Loss':<10} | {'Val Precision':<10} | {'Val Recall':<10} | {'Best Epoch':<10}")
        print("-" * 65)
        for model_name, results in model_ablation_results.items():
            if "error" in results:
                print(f"{model_name:<25} | {'ERROR':<10} | {'-':<10} | {'-':<10}")
            else:
                print(f"{model_name:<25} | {results.get('val_dice', -1):<10.4f} | "
                      f"{results.get('val_iou', -1):<10.4f} | "
                      f"{results.get('val_loss', -1):<10.4f} | "
                      f"{results.get('val_precision', -1):<10.4f} | "
                      f"{results.get('val_recall', -1):<10.4f} | "
                      f"{results.get('best_epoch', 'N/A'):<10}")

if __name__ == "__main__":
    try:
        from PIL import Image
    except ImportError:
        print("Error: Pillow (PIL) library not found. Please install it: pip install Pillow")
        exit()
    RUN_ABLATION_STUDY = True 
    RUN_FULL_TRAINING = True
    
    main()
```
